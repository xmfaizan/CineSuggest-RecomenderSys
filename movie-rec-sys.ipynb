```python
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
def dict_to_list(obj):
    list=[]
    for i in ast.literal_eval(obj):
        list.append(i['name'])
    return list    
```

```python
movies['genres'] = movies['genres'].apply(dict_to_list)
```

```python
movies.head()
```

```python
movies.iloc[0].keywords
```

```python
movies['keywords'] = movies['keywords'].apply(dict_to_list)
```

```python
movies.head()
```

```python
movies['cast'][0]  #Pick the first three names of the actors
```

```python
def dict_to_list3(obj):
    list=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter != 3:
            list.append(i['name'])
            counter+=1
        else:
            break
    return list    
```

```python
movies['cast'] = movies['cast'].apply(dict_to_list3)
```

```python
movies['cast']
```

```python
movies.head()
```

```python
movies['crew'][0] #Pick the director name
```

```python
def fetch_dir(obj):
    list=[]
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            list.append(i['name'])
            break
    return list        
```

```python
movies['crew'] = movies['crew'].apply(fetch_dir)
```

```python
movies['crew']
```

```python
movies.head()
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split()) #Splitting overview into list
```

```python
movies['overview']
```

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['cast'] + movies['crew']
```

```python
movies['tags']
```

```python
updated_df = movies[['movie_id','title','tags']]
```

```python
updated_df
```

```python
updated_df['tags'] = updated_df['tags'].apply(lambda x:" ".join(x))
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,ngram_range=(1,2),stop_words='english')
```

```python
vec = cv.fit_transform(updated_df['tags']).toarray()
```

```python
import nltk 
from nltk.stem import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
ps.stem('caring')
```

```python
def stem(txt):
    y=[]
    for i in txt.split():
        y.append(ps.stem(i))
    return " ".join(y)     
```

```python
updated_df['tags'] = updated_df['tags'].apply(stem)
```

```python
def recommend(movie):
    movie_index = updated_df[updated_df['title'] == movie].index[0]
    distances = sim[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    for i in movies_list:
        print(updated_df.iloc[i[0]].title)
```

```python
recommend('The Dark Knight Rises')
```

```python
recommend('Vanilla Sky')
```
